# Matplotlib Basics
Matplotlib makes data visualizations very easy.

## Simple graphs

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 10, 0.1) # plot points every 0.1 from 0 to 10 
```

```python
y1 = np.sin(x)
```

```python
plt.plot(x,y1)
```

```python
plt.show()
```

## Multiple lines in one graph

```python
y2 = np.cos(x)
```

```python
plt.plot(x, y1, label='sin')
plt.plot(x, y2, linestyle='--', label='cos')
plt.xlabel('x') # you can name this anything, like 'foo' or 'bar'
plt.ylabel('y') # you can name this anything, like 'foo' or 'bar'
plt.title('sin & cos')
plt.legend()
plt.show()
```
